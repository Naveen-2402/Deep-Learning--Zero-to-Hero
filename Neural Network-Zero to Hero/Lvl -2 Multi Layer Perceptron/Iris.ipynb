```python
# Types of Iris is 3
# Its not Binary so Multi Layer Perceptron
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Iris.csv')
#df = df.sample(frac=1)     #suffle randomly
df.head(10)
```

```python
# One-hot Encoding
y = pd.get_dummies(df["Species"], dtype=int)
y
```

```python
x = df.drop (["Id", "Species"], axis=1) # axis=1, remove by column
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
# Creating Multi Layer Perceptron Model
# pip install keras  
# pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
"""
keras.layers.Dense(
    units,
    activation=None,
    use_bias=True,
    kernel_initializer="glorot_uniform",
    bias_initializer="zeros",
    kernel_regularizer=None,
    bias_regularizer=None,
    activity_regularizer=None,
    kernel_constraint=None,
    bias_constraint=None,
    lora_rank=None,
    **kwargs
)
"""
# Paramets for Dense
```

```python
model = Sequential()
model.add(Dense(6, activation="sigmoid"))    # Hidden layer with 6 nodes 
model.add(Dense(3, activation="softmax"))    # Output layer

model.compile(loss="categorical_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=50, batch_size=5)
```

```python
model.evaluate(X_test, y_test)
```

```python
import numpy as np

# Class names corresponding to the target labels
class_names = ["setosa", "versicolor", "virginica"]

input_features = np.array([[5.1, 3.5, 1.4, 0.2]])  # Convert list to NumPy array
pred = model.predict(input_features)

# Get the index of the highest probability
predicted_class_index = np.argmax(pred, axis=1)

predicted_class_name = class_names[predicted_class_index[0]]

# Print the results
print("Prediction Probabilities:", pred)
print("Predicted Class Index:", predicted_class_index)
print("Predicted Iris Type:", predicted_class_name)
```
